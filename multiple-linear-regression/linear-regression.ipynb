```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
class LinearRegression:
    def __init__(self, X, Y, lr=0.0001, max_iter=10000):
        self.x = X
        self.y = Y.reshape(-1, 1)
        self.max_iter = max_iter
        self.lr = lr
    

    def fit(self):
        w = np.zeros((self.x.shape[1], 1))        
        b = 0        

        for _ in range(self.max_iter):            
            pred = np.dot(self.x, w.T) + b        

            error = np.mean((pred - self.y))            

            dJ_dw = np.mean(self.x * (pred - self.y), axis=0)
            dJ_db = error    
            
            w = w - self.lr * dJ_dw
            b = b - self.lr * dJ_db 
        
        self.w = w
        self.b = b

    def predict(self, x):        
        return (np.dot(x, self.w.T) + self.b).flatten()
    
```

```python
X = np.array([x for x in range(0, 100, 5)])
y = np.array([4 * x + np.random.randint(0, 100) for x in X])
```

```python
plt.plot(X, y, 'o')
```

```python
X_train = X.reshape(-1, 1)
y_train = y.reshape(-1, 1)
```

```python
model = LinearRegression(X_train, y_train, lr=0.0001, max_iter=100000)
model.fit()
model.w, model.b
```

```python

plt.plot(X, y, 'o')
plt.plot(X, model.predict(X.reshape(-1, 1)), 'r')
```
