```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,  MinMaxScaler
```

```python
data = pd.read_csv("./healthcare-dataset-stroke-data.csv")
data.replace(['Unknown', 'None', 'N/A', 'NaN', 'NULL', '?',''], pd.NA, inplace=True)
```

```python
nulls = data.isna().sum()
print("Number of null values", nulls)
```

# Handling Missing Values

### Drop the row of a null value

```python
data.dropna()
```

### Forward and Backward fill of the null values

```python
data.fillna(method="ffill")
```

```python
data.fillna(method="bfill")
```

### Fill with the columns mean or median or mode

```python
data.fillna(data["bmi"].mean())
```

```python
data.fillna(data["bmi"].median())
```

```python
data.fillna(data["bmi"].mode())
```

### Drop rows of the data whose target(stroke) values are missing (General recommended approach)

```python
data.dropna(subset=["stroke"], inplace=True)
```

### Divide the dataset into Dependent & Independent variable

```python
X = data.iloc[:, :-1]
Y = data.iloc[:, -1:]
```

```python
num_cols = ['age', 'avg_glucose_level', 'bmi']
categorical_cols = ['gender', 'hypertension', 'heart_disease', 'smoking_status', ]
```

### KNN algorithm to fill the missing values (only numerical columns)

```python
KNN_imputer = KNNImputer(n_neighbors = 3)
scaled_cols = pd.DataFrame(KNN_imputer.fit_transform(X[num_cols]),columns=num_cols)

scaled_cols.info()
```

### Most frequent mode to fill missing values (only categorical values)

```python
categorical_cols = ['gender', 'hypertension', 'heart_disease', 'smoking_status']
categorical_imputer = SimpleImputer(strategy='most_frequent',  missing_values=pd.NA)

categorical_imputer. fit(X[categorical_cols])
X_categorical_imputed = pd.DataFrame(categorical_imputer.transform(X[categorical_cols]), columns=categorical_cols)

X_categorical_imputed.info()

```

# Handling Outliers

```python
 X = pd.concat([scaled_cols, X_categorical_imputed], axis=1)
```

### Distrubtion of all columns potentially have outliers

```python
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
sns.histplot(X["age"])

plt.subplot(1, 3, 2)
sns.histplot(X["avg_glucose_level"])

plt.subplot(1, 3, 3)
sns.histplot(X["bmi"])

plt.show()
```

## Age is normaly distrubited  while avg_glucose_level and bmi are left skewed

#### For age use Z-score treatment

```python
X["age"].describe()
```

```python
upper_limit = round(X["age"].mean() + 1.5 * X["age"].std(), 4)
lower_limit = round(X["age"].mean() - 1.5 * X["age"].std(), 4)
print("upper limit ", upper_limit, " lower limit ", lower_limit)
```

```python
outliers = X[(X["age"] < lower_limit) | (X["age"] > upper_limit)]
outliers
```

```python
trimmed_X = X[(X["age"] >= lower_limit) & (X["age"] <= upper_limit)]
trimmed_X
```

```python
X["age"] = np.where(X["age"] < lower_limit, lower_limit, X["age"])
X["age"] = np.where(X["age"] > upper_limit, upper_limit, X["age"])

X["age"].describe()
```

#### For avg_glucose_level Box Method

```python
X["avg_glucose_level"].describe()
```

```python
sns.boxplot(X["avg_glucose_level"])
```

```python
Q1 = X["avg_glucose_level"].quantile(0.25)
Q3 = X["avg_glucose_level"].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR
print("upper limit ", upper_limit, " lower limit ", lower_limit)
```

```python
outliers = X[(X["avg_glucose_level"] < lower_limit) | (X["avg_glucose_level"] > upper_limit)]
outliers
```

```python
trimmed_X = X[(X["avg_glucose_level"] >= lower_limit) & (X["avg_glucose_level"] <= upper_limit)]
trimmed_X["avg_glucose_level"].describe()
```

```python
X["avg_glucose_level"] = np.where(X["avg_glucose_level"] < lower_limit, lower_limit, X["avg_glucose_level"])
X["avg_glucose_level"] = np.where(X["avg_glucose_level"] > upper_limit, upper_limit, X["avg_glucose_level"])

sns.boxplot(X["avg_glucose_level"])
```

#### For bmi Box Method

```python
X["bmi"].describe()
```

```python
sns.boxplot(X["bmi"])
```

```python
Q1 = X["bmi"].quantile(0.25)
Q3 = X["bmi"].quantile(0.75)
IQR = round(Q3 - Q1, 4)
print(IQR)
```

```python
upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR
print("upper limit ", upper_limit, " lower limit ", lower_limit)
```

```python
outliers = X[(X["bmi"] < lower_limit) | (X["bmi"] > upper_limit)]
outliers
```

```python
trimmed_X = X[(X["bmi"] >= lower_limit) & (X["bmi"] <= upper_limit)]
trimmed_X["bmi"].describe()
```

```python
X["bmi"] = np.where(X["bmi"] < lower_limit, lower_limit, X["bmi"])
X["bmi"] = np.where(X["bmi"] > upper_limit, upper_limit, X["bmi"])

sns.boxplot(X["bmi"])
```
